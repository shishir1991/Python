```python
#3. The following table shows the mean and variance of the expenditure for two groups of people. 
#You want to compare the variability in expenditure for both groups with respect to their mean. 
#Which statistical measure would you use to evaluate the variability in expenditure? 
#Please provide an explanation for your answer.
#S.No       Expenditure            Group 1        Group 2
#           Mean                  $500,000       $40,000
#           Standard Dev.         $125,000       $10,000

To compare the variability in expenditure for both groups with respect to their mean, you would use the coeff. of variation (CV).
Coefficient of variance (CV) = (Standard dev. / Mean)  * 100%
For
Group 1: CV = (125,000 / 500,000) * 100% = 25%
Group 2: CV = (10,000 / 40,000) * 100% = 25%
Both groups have the same coefficient of variation (25%), 
indicating that the variability in expenditure relative to the mean is consistent across both groups.
```

```python
#4. During the survey, the ages of 80 patients infected by COVID and admitted to one of the city hospitals
#were recorded and the collected data is represented in the less than cumulative frequency distribution table.

#Age (in yrs.)         No. of Patients
#  5 - 15                    6
# 15 - 25                    11
# 25 - 35                    21
# 35 - 45                    23
# 45 - 55                    14
# 55 - 65                    5

#a. Which class interval has the highest frequency?
#b. Which age was affected the least?
#c. How many patients of age 45 years and above were admitted?
#d. Which is the modal class interval in the above dataset
#e. What is the median class interval of age?

A. The class interval with the highest frequency is "35-45 years," with 23 patients.

B. As the age group “55-65 years” have only 5 patients. 
   Therefore, this age range was affected the least among the patients admitted to the hospital.
    
C. Since the interval spans 10 years (from 35 to 45), and assuming a uniform distribution, 
   we can estimate that approximately 1/10th of the patients in this interval are exactly 45 years old.
   So, for the "35 to 45 years" interval: 
   Number of patients aged 45 = (1/10) * 23 = 2.3, rounding off 2
   Total patients of age 45 years and above
   14 (from "45 to 55 years") + 2 (estimated from "35 to 45 years") + 5 (from "55 to 65 years") = 21
   So, there were approximately 21 patients of age 45 years and above admitted to the hospital.
    
D. The modal class interval is the interval "35 to 45 years" because it has the highest frequency of 23 patients.

E. To find the median class interval, we locate the class interval containing the median age, 
   which is the interval where the cumulative frequency first exceeds (or equals) half of the total frequency.
   Cumulative Frequency:
    5 to 15 years: 6
   15 to 25 years: 6 + 11 = 17
   25 to 35 years: 17 + 21 = 38
   35 to 45 years: 38 + 23 = 61
   45 to 55 years: 61 + 14 = 75
   55 to 65 years: 75 + 5 = 80
   Since the total frequency is 80, the median would be the 40th value. Looking at the cumulative frequencies,
   the interval "35 to 45 years" (from 38 to 61) contains the 40th value.
   So, the median class interval of age is "35 to 45 years."
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Prob.9-> Evaluate the measures of central tendency and measures of dispersion for all the quantitative variables in the dataset.
```

```python
# Measures of central tendency & dispersion
central_tendency = bank_loan_data.select_dtypes(include=['int64', 'float64']).agg(['mean', 'median'])
dispersion = bank_loan_data.select_dtypes(include=['int64', 'float64']).agg(['std', 'min', 'max', lambda x: x.quantile(0.25), lambda x: x.quantile(0.75)])
summary_stats = pd.concat([central_tendency, dispersion], axis=0)
summary_stats.index = ['Mean', 'Median', 'Standard Deviation', 'Minimum', 'Maximum', '1st Quartile', '3rd Quartile']
summary_stats
```

```python
# Prob. 10-->What statistical method will you use to examine the presence of a linear relationship between age and experience
#variables? Also, create a plot to illustrate this relationship.
```

```python
#To examine the presence of a linear relationship between the "age" and "experience" variables, we can use correlation coefficient.
#It measures the strength and direction of the linear relationship between two continuous variables.
#If the correlation coefficient is close to 1 or -1, it indicates a strong linear relationship,
#while values close to 0 indicate little to no linear relationship.
correlation_coefficient = bank_loan_data[['Age', 'Experience']].corr().iloc[0, 1]
correlation_coefficient
```

```python
# A scatter plot to illustrate the relationship between age and experience
plt.figure(figsize=(8, 6))
sns.scatterplot(data=bank_loan_data, x="Age", y="Experience")
plt.title(f"Scatter Plot of Age vs Experience\nCorrelation Coefficient: {correlation_coefficient:.2f}")
plt.xlabel("Age")
plt.ylabel("Experience")
plt.grid(True)
plt.show()
```

```python
#Prob.13--> The ‘Mortgage’ variable has a lot of zeroes. Impute with some business logical value that you feel fit for the data.
```

```python
# Since the variable has a lot of zeroes, indicating that many individuals have no mortgage.
# We will impute the zeroes with the median value of 'Mortgage' for those individuals who have a non-zero mortgage.
nz_median_mortgage = bank_loan_data[bank_loan_data['Mortgage'] > 0]['Mortgage'].median()
print("Median of mortgage is:",nz_median_mortgage)
```

```python
# Impute zeroes in the 'Mortgage' variable with the median value for non-zero mortgage holders
bank_loan_data['Mortgage'] = bank_loan_data['Mortgage'].replace(0, nz_median_mortgage)

# head is ued Display the first few rows of the updated dataset to verify the imputation
bank_loan_data.head()
```

```python
#Prob.19--> How many customers use online banking? Do customers using bank internet facilities have higher income?
```

```python
# Count the number of customers who use online banking
online_cust_count = bank_loan_data['Online'].sum()
print("Number of customers using online banking facilities:", online_cust_count)
```

```python
# Now compare income distribution for customers who use online banking vs. those who don't
online_income = bank_loan_data[bank_loan_data['Online'] == 1]['Income']
offline_income = bank_loan_data[bank_loan_data['Online'] == 0]['Income']
# We'll now calculate the summary statistics such as mean or median income for both groups and compare them.
online_mean_income = online_income.mean()
offline_mean_income = offline_income.mean()
print("Mean income for customers using online banking:", online_mean_income)
print("Mean income for customers not using online banking:", offline_mean_income)
```

```python
#Prob.20--> Using the z-score of the income variable, find out the number of observations outside the +-3σ.
```

```python
from scipy.stats import zscore

# Calculate z-scores for the 'Income' variable
income_z_scores = zscore(bank_loan_data['Income'])
income_z_scores
```

```python
# Count the number of observations outside the range of ±3 standard deviations
outliers_count = sum((income_z_scores > 3) | (income_z_scores < -3))
print("Number of observations outside the range of ±3 standard deviations from the mean:", outliers_count)
```
